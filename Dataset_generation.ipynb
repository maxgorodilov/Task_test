```python
import pandas as pd
import numpy as np

# Установим параметры
num_records = 1_000_0000
num_duplicates = int(num_records * 0.1)  # 10% дублей
null_fraction = 0.1  # Доля NULL значений (10%)

# Генерация данных
def generate_data(num_records):
    # Генерация числовых данных
    numeric_data = np.random.rand(num_records) * 1000  # случайные числа от 0 до 1000

    # Генерация дат
    start_date = np.datetime64('2010-01-01')
    end_date = np.datetime64('2023-01-01')
    date_range = (end_date - start_date).astype('timedelta64[D]').astype(int)

    random_days = np.random.randint(0, date_range + 1, size=num_records)
    random_dates = start_date + random_days.astype('timedelta64[D]')

    # Генерация случайного времени
    random_hours = np.random.randint(0, 24, size=num_records)
    random_minutes = np.random.randint(0, 60, size=num_records)
    random_seconds = np.random.randint(0, 60, size=num_records)

    random_times = (random_hours * 3600 + random_minutes * 60 + random_seconds).astype('timedelta64[s]')
    datetime_data = random_dates + random_times

    # Генерация строковых данных
    string_data = ['item_' + str(i) for i in range(num_records)]

    # Создание DataFrame
    df = pd.DataFrame({
        'numeric_column': numeric_data,
        'datetime_column': datetime_data,
        'string_column': string_data
    })

    return df

# Генерация основного датасета
df = generate_data(num_records)

# Добавление NULL значений
num_nulls = int(num_records * null_fraction)
for column in ['numeric_column', 'datetime_column', 'string_column']:
    null_indices = np.random.choice(df.index, num_nulls, replace=False)
    df.loc[null_indices, column] = np.nan

# Добавление дублей
duplicates = df.sample(num_duplicates, replace=True)
df_with_duplicates = pd.concat([df, duplicates], ignore_index=True)

# Перемешивание строк
df_with_duplicates = df_with_duplicates.sample(frac=1).reset_index(drop=True)

# Сохранение в CSV файл
df_with_duplicates.to_csv('test_dataset.csv', index=False)
from google.colab import files
files.download('test_dataset.csv')
```
